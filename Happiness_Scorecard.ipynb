```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Define the data as a dictionary
data = {
    'Neha': [6, 7, 6, 7, 9, 8, 9, 8, 8, 8, 10],
    'Amit': [8, 10, 9, 9, 9, 10, 9, 9, 9, 8, 8],
    'Sachin': [7, 7, 7, 10, 10, 10, 9, 9, 8, 9, 9]
}

# Convert data to a pandas DataFrame
df = pd.DataFrame(data)

# Calculate the mean and standard deviation for each column
means = df.mean()
stds = df.std()

# Set the bar width
bar_width = 0.2

# Create the bar chart
fig, ax = plt.subplots()
index = np.arange(len(means))
rects1 = ax.bar(index - bar_width, means, bar_width, yerr=stds, label='Neha')
rects2 = ax.bar(index, means, bar_width, yerr=stds, label='Amit')
rects3 = ax.bar(index + bar_width, means, bar_width, yerr=stds, label='Sachin')

# Add x-axis labels and title
ax.set_xlabel('Employee')
ax.set_ylabel('Happiness Score')
ax.set_title('Weekly Scores')

# Add legend and adjust layout
ax.legend()
fig.tight_layout()

# Show the plot
plt.show()
```
