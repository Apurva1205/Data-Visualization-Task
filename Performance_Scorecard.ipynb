```python
import matplotlib.pyplot as plt

# define the data as a dictionary with keys for each team member
data = {
    "Neha": [8.67, 7.67, 8.33],
    "Amit": [7.67, 7.33, 7.0],
    "Sachin": [7.67, 6.67, 7.0]
}

# define the x-axis labels (weeks 1, 2, and 3)
labels = ["Week 1", "Week 2", "Week 3"]

# create a line chart with a line for each team member
for member, scores in data.items():
    plt.plot(labels, scores, label=member)

# add a legend to the chart
plt.legend()

# add axis labels and a title
plt.xlabel("Week")
plt.ylabel("Performance Score")
plt.title("Performance Scorecard")

# display the chart
plt.show()
```
