```python
import matplotlib.pyplot as plt

# Define the data for each parameter
weeks = ['Week 1', 'Week 2', 'Week 3']
param1_neha = [10, 0, 75]
param1_amit = [8, 0, 100]

# Create the bar chart for Parameter 1
plt.bar(weeks, param1_neha, label='Neha')
plt.bar(weeks, param1_amit, label='Amit')
plt.title('Parameter 1: Efforts put into completing projects')
plt.xlabel('Weeks')
plt.ylabel('Score')
plt.legend()
plt.show()
```
